```python
#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('HousePrices.csv')
```

```python
dataset=dataset.dropna(how='any',axis=1)
```

```python
dataset=dataset.dropna(how='any',axis=0)
```

```python
#extract categorical var
categorical=[var for var in dataset.columns if dataset[var].dtype=='O']
```

```python
categorical
```

```python
dataset[categorical].head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
dataset['BldgType']=label_encoder.fit_transform(dataset["BldgType"])
dataset['HouseStyle']=label_encoder.fit_transform(dataset["HouseStyle"])
dataset['Street']=label_encoder.fit_transform(dataset["Street"])
dataset['Foundation']=label_encoder.fit_transform(dataset["Foundation"])
dataset['HeatingQC']=label_encoder.fit_transform(dataset["HeatingQC"])
dataset['CentralAir']=label_encoder.fit_transform(dataset["CentralAir"])
dataset['SaleCondition']=label_encoder.fit_transform(dataset["SaleCondition"])
dataset['SaleType']=label_encoder.fit_transform(dataset["SaleType"])
```

```python
#extract numerical var
numerical=[var for var in dataset.columns if dataset[var].dtype!='O']
```

```python
numerical
```

```python
dataset=dataset[numerical]
dataset.head()
```

```python
dataset=dataset.drop(['Id'],axis=1)
```

```python
dataset.head()
```

```python
X = dataset.iloc[:,:-1]
X.head()
```

```python
y=dataset.iloc[:,-1]
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
#checking the score
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
